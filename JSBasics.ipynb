```python
//clasically everything is described using vars in javascript
var myName = 'Bram';
console.log(myName);

//variables can be changed
myName = 'John';
console.log(myName);

//in newer versions, let replaces var
let myName = 'Bram';
console.log(myName);


myName = 'John';
console.log(myName);

//const produces constants
const myName = 'Bram';
console.log(myName);

//const values cannot be changed, running the code below produces errors
myName = 'John';
console.log(myName);
```

```python
//Regular JS functions work like this
function printMyName(name) {
    console.log(name);
}
printMyName('Bram')

//Arrow functions work like this
const printMyName = (name) => {
    console.log(name)
}
printMyName('Bram')

//Empty arrow function
const printMyName = () => {
    console.log('Bram')
}
printMyName()

//Multiple arguments
const printMyName = (name, age) => {
    console.log(name, age)
}
printMyName('Bram', 28)

//Compact return
const multiply = number => number * 2
console.log(multiply(2))
```
